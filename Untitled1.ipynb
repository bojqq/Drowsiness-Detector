# Step 4:  Analyze the relationship between audio features and popularity  
We calculate correlations between popularity and audio features to see which features are most related.



## Goal 1 - Relationship between audio features and popularity  
We investigate how audio features such as danceability, energy, valence, and tempo  
relate to the popularity of songs.  

The approache used:  
- **Averages:** Compare features between high-popularity and low-popularity songs.  



```python
# Split songs into high and low popularity
df['popularity_group'] = np.where(df['popularity'] >= 50, 'High', 'Low')

# Calculate average features per group
avg_features = df.groupby('popularity_group')[['danceability','energy','valence','tempo']].mean()
print(avg_features)

# Plot averages

avg_features.T.plot(kind='bar', figsize=(8,5))
plt.title("Average Audio Features by Popularity Group")
plt.ylabel("Average Value")
plt.xlabel("Audio Features")
plt.ylim(0.4, 0.7)  # Zoom in to highlight differences
plt.legend(title="Popularity Group")
plt.grid(True)
plt.show()
```



### Key Results:
- High popularity songs show slightly lower **energy** and **valence** compared to low popularity songs.
- Differences in **danceability** and **tempo** between the two groups are minimal.
- This indicates that popular songs are not necessarily more energetic or more positive in mood.

### Conclusion:
Popularity does not strongly depend on energy or tempo. Other factors such as lyrics, artist identity, or external promotion may play a bigger role in determining whether a song becomes popular.
